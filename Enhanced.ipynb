```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('don.jpeg')

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.show()

# Apply Gaussian Blur
blur = cv2.GaussianBlur(gray, (11, 11), 0)
plt.imshow(blur, cmap='gray')
plt.title("Blurred Image")
plt.show()

# Apply Canny Edge Detection
canny = cv2.Canny(blur, 30, 150, 3)
plt.imshow(canny, cmap='gray')
plt.title("Canny Edge Detection")
plt.show()

# Dilate edges to close gaps
kernel = np.ones((3,3), np.uint8)  # Structuring element
dilated = cv2.dilate(canny, kernel, iterations=1)
plt.imshow(dilated, cmap='gray')
plt.title("Dilated Image")
plt.show()

# Find contours
cnt, _ = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# Draw contours on the original image
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0, 255, 0), 2)

# Display the result
plt.imshow(rgb)
plt.title("Detected Objects")
plt.show()

# Print the object count
print("Objects in the image:", len(cnt))
```
